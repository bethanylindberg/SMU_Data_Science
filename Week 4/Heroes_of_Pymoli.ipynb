# Data Analysis

```python
# Dependencies and Setup
import pandas as pd
from pandas import DataFrame
```

```python
gamefile = pd.read_csv('C:\\Users\\bethf\\Desktop\\Resources\\purchase_data.csv')

gamedata = DataFrame(gamefile)

gamedata.head()
```

Player Count

```python
players_summary = gamedata.groupby(["SN"])["SN"].value_counts()

players_summary_total = DataFrame({"Total Players":players_summary.count()},index=[0])

players_summary_total
```

Purchasing Analysis (Total)

```python
purchasing_analysis_total_summary_table = DataFrame({"Number of Unique Items":gamedata["Item ID"].nunique(),\
                                                     "Average Price":gamedata["Price"].mean(),\
                                                    "Number of Purchases":gamedata["Purchase ID"].count(),\
                                                     "Total Revenue":gamedata["Price"].sum()},index=[0])
purchasing_analysis_total_summary_table["Average Price"] = purchasing_analysis_total_summary_table["Average Price"].map("${:.2f}".format)
purchasing_analysis_total_summary_table["Total Revenue"] = purchasing_analysis_total_summary_table["Total Revenue"].map("${:.2f}".format)
purchasing_analysis_total_summary_table
```

Gender Demographics

```python
genders = gamedata.drop_duplicates(subset=["SN"])["Gender"].value_counts()

gender_demographic_summary_table = DataFrame({"Total Count":genders})

gender_demographic_summary_table["Percentage of Players"] = gender_demographic_summary_table["Total Count"] \
                                    / gender_demographic_summary_table["Total Count"].sum() * 100

gender_demographic_summary_table["Percentage of Players"] = gender_demographic_summary_table["Percentage of Players"].map("{:.2f}%".format)
gender_demographic_summary_table
```

Purchasing Analysis (Gender)

```python
purchasing_analysis_gender_summary_table = DataFrame({"Purchase Count":gamedata.groupby("Gender")["Purchase ID"].count(),\
                                                      "Average Purchase Price":gamedata.groupby("Gender")["Price"].mean(),\
                                                     "Total Purchase Value":gamedata.groupby("Gender")["Price"].sum(),\
                                                     "Average Purchase per person":gamedata.groupby("Gender")["Price"].sum()/genders})

purchasing_analysis_gender_summary_table["Average Purchase Price"] = purchasing_analysis_gender_summary_table["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_gender_summary_table["Total Purchase Value"] = purchasing_analysis_gender_summary_table["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis_gender_summary_table["Average Purchase per person"] = purchasing_analysis_gender_summary_table["Average Purchase per person"].map("${:.2f}".format)
purchasing_analysis_gender_summary_table
```

Age Demographics

```python
bins = [0, 9, 14, 19, 24, 29, 34,39,999] # Make Bins for ages
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39", "40+"] #Make Names for age groups
gamedata2 = gamedata #Copy data frame to preserve original
 # Slice the data and place it into bins
gamedata2["Age Groups"] = pd.cut(gamedata["Age"], bins, labels=group_labels)

ages = gamedata.drop_duplicates(subset=["SN"])["Age Groups"].value_counts()

age_demographic_summary_table = DataFrame({"Total Count":ages})

age_demographic_summary_table["Percentage of Players"] = age_demographic_summary_table["Total Count"] \
                                    / age_demographic_summary_table["Total Count"].sum() * 100
age_demographic_summary_table["Percentage of Players"] = age_demographic_summary_table["Percentage of Players"].map("{:.2f}%".format)
age_demographic_summary_table_sorted = age_demographic_summary_table.sort_index()

age_demographic_summary_table_sorted
```

Purchasing Analysis (Age)

```python
age_purchasing_summary_table = DataFrame({"Purchase Count":gamedata2.groupby("Age Groups")["Purchase ID"].count(),\
                                            "Average Purchase Price":gamedata2.groupby("Age Groups")["Price"].mean(),\
                                            "Total Purchase Value":gamedata2.groupby("Age Groups")["Price"].sum(),\
                                            "Average Purchase per person":gamedata2.groupby("Age Groups")["Price"].sum()/ages})

age_purchasing_summary_table["Average Purchase Price"] = age_purchasing_summary_table["Average Purchase Price"].map("${:.2f}".format)
age_purchasing_summary_table["Total Purchase Value"] = age_purchasing_summary_table["Total Purchase Value"].map("${:.2f}".format)
age_purchasing_summary_table["Average Purchase per person"] = age_purchasing_summary_table["Average Purchase per person"].map("${:.2f}".format)

age_purchasing_summary_table
```

Top Spenders

```python
top_spenders_summary_table = DataFrame({"Purchase Count":gamedata.groupby("SN")["Purchase ID"].count(),\
                                            "Average Purchase Price":gamedata.groupby("SN")["Price"].mean(),\
                                            "Total Purchase Value":gamedata.groupby("SN")["Price"].sum()})

top_spenders_summary_table_sorted = top_spenders_summary_table.sort_values(["Total Purchase Value"], ascending=False)

top_spenders_summary_table_sorted["Average Purchase Price"] = top_spenders_summary_table_sorted["Average Purchase Price"].map("${:.2f}".format)
top_spenders_summary_table_sorted["Total Purchase Value"] = top_spenders_summary_table_sorted["Total Purchase Value"].map("${:.2f}".format)
top_spenders_summary_table_sorted.head()
```

Most Popular Items

```python
most_popular_items = gamedata.groupby(["Item ID","Item Name"])      

most_popular_items_summary_table = DataFrame({"Purchase Count":most_popular_items["Purchase ID"].count(),\
                                           "Item Price":most_popular_items["Price"].max(),\
                                           "Total Purchase Value":most_popular_items["Price"].sum()})

most_popular_items_summary_table_sorted = most_popular_items_summary_table.sort_values("Purchase Count", ascending=False)

most_popular_items_summary_table_sorted["Item Price"] = most_popular_items_summary_table_sorted["Item Price"].map("${:.2f}".format)
most_popular_items_summary_table_sorted["Total Purchase Value"] = most_popular_items_summary_table_sorted["Total Purchase Value"].map("${:.2f}".format)
most_popular_items_summary_table_sorted.head(10)
```

 Most Profitable Items

```python
most_profitable_items_summary_table_sorted = most_popular_items_summary_table.sort_values("Total Purchase Value", ascending=False)

most_profitable_items_summary_table_sorted["Item Price"] = most_profitable_items_summary_table_sorted["Item Price"].map("${:.2f}".format)
most_profitable_items_summary_table_sorted["Total Purchase Value"] = most_profitable_items_summary_table_sorted["Total Purchase Value"].map("${:.2f}".format)
most_profitable_items_summary_table_sorted.head(25)
```

Most profitable share of revenue

```python
most_profitable_items_sum10 = most_popular_items_summary_table["Total Purchase Value"].nlargest(n=10, keep='first').sum()
most_profitable_items_sum25 = most_popular_items_summary_table["Total Purchase Value"].nlargest(n=25, keep='first').sum()
revenue = gamedata["Price"].sum()
most_profitable_share_of_total = DataFrame({"Top Ten Revenue":most_profitable_items_sum10,\
                                            "Percentage of total(10)":most_profitable_items_sum10/revenue*100,\
                                            "Top Twenty-Five Revenue":most_profitable_items_sum25,\
                                          "Percentage of total(25)":most_profitable_items_sum25/revenue*100},index=[0])

most_profitable_share_of_total["Top Ten Revenue"] = most_profitable_share_of_total["Top Ten Revenue"].map("${:.2f}".format)
most_profitable_share_of_total["Percentage of total(10)"] = most_profitable_share_of_total["Percentage of total(10)"].map("{:.2f}%".format)
most_profitable_share_of_total["Top Twenty-Five Revenue"] = most_profitable_share_of_total["Top Twenty-Five Revenue"].map("${:.2f}".format)
most_profitable_share_of_total["Percentage of total(25)"] = most_profitable_share_of_total["Percentage of total(25)"].map("{:.2f}%".format)
most_profitable_share_of_total
```
